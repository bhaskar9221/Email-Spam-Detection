#Email Spam Detection

##Importing our necessary libraries.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import sklearn
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning Projects/Email Spam Detection/mail_dataset.csv")
```

```python
print(df)
```

```python
data = df.where((pd.notnull(df)),'')
```

```python
data.loc[data['Category'] == 'spam', 'Category',] = 0
data.loc[data['Category'] == 'ham', 'Category',]  = 1
```

```python
X = data['Message']
Y = data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df =1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features  = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test  = Y_test.astype('int')
```

```python
print(X_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
predirction_on_trainData = model.predict(X_train_features)
accuracy_on_trainingData = accuracy_score(Y_train,predirction_on_trainData) 
```

```python
print('Accuracy on the training data : ',accuracy_on_trainingData)
```

```python
prediction_on_testData = model.predict(X_test_features)
accuracy_on_testData   = accuracy_score(Y_test,prediction_on_testData)
```

```python
print('Accuracy on the training data : ',accuracy_on_testData)
```

```python
customInput = ["Congratulations! You've been selected as a lucky winner of our exclusive prize. Claim your reward now and start enjoying the benefits!"]
input_data_features = feature_extraction.transform(customInput)
prediction = model.predict(input_data_features)
print(prediction)


if(prediction[0] == 1):
  print('Ham Mail')
else:
  print('Spam Mail')
```
